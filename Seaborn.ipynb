# Seaborn

Seaborn is a library for making statistical graphics in Python. It is built on top of matplotlib and closely integrated with pandas data structures.

```python
import seaborn as sns
```

```python
# Let's use a built in dataset to demonstrate seaborn
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
# The function relplot() is named that way because it is designed to visualize many different statistical relationships. 
```

This particular plot shows the relationship between five variables in the tips dataset. Three are numeric, and two are categorical. 

* Two numeric variables **(total_bill and tip)** determined the position of each point on the axes.
* Third **(size)** determined the size of each point. 
* One categorical variable **(time)** split the dataset onto two different axes (facets). 
* And the other **(smoker)** determined the color and shape of each point.

```python

sns.relplot(x="total_bill",y="tip",hue="smoker",style="smoker", kind="line",col="time",data=tips)
```

**Statistical estimation in seaborn goes beyond descriptive statisitics. For example, it is also possible to enhance a scatterplot to include a linear regression model (and its uncertainty) using lmplot():**

```python
sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",
           data=tips);
```

Standard scatter and line plots visualize relationships between numerical variables, but many data analyses involve **categorical variables** . There are several specialized plot types in seaborn that are optimized for visualizing this kind of data. They can be accessed through **catplot()**. Similar to **relplot()**, the idea of **catplot()** is that it exposes a common dataset-oriented API that generalizes over different representations of the relationship between one numeric variable and one (or more) **categorical variables.**

we can use kernel density estimation to represent the underlying distribution that the points are sampled from:

```python
tips = sns.load_dataset("tips")
tips.head()
sns.catplot(x="day", y="total_bill",data=tips,kind="violin",hue="smoker",split=True)
```

### joinplot()
It can be used to make visualizations with multiple plots.

```python
iris=sns.load_dataset("iris")
print(iris.head())

sns.jointplot(x="sepal_length",y="petal_length",data=iris)
```

####  As we can see above, both scatterplot and histograms are used.

## pairplot()

It takes a broader view, showing all pairwise relationships and the marginal distributions.

```python
sns.pairplot(data=iris,hue="species")
```
