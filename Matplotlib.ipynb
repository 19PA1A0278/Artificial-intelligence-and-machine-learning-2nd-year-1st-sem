# matplotlib
Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. 

Matplotlib tries to make easy things easy and hard things possible. You can generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc., with just a few lines of code.

```python
# Import the required libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set a seed. This is to make sure that everyone has same set of random numbers.
np.random.seed(0)

# To make sure that the output is displayed on jupyter notebook without calling plt.show() everytime
%matplotlib inline
```

### Two plotting approaches
There are two approaches to plotting the:

* The functional approach
* Object-oriented approach (preferred)

#  The Functional approach

will create a figure object ( the canvas that we plot on ) and an axes object that is the plot with lines and info. 

### 1) Create an empty plot

```python
plt.plot()
```

### 2) Plot the given points in list

```python
x=[1,2,3]
y=[11,17,35]
plt.plot(x,y)
```

### 3) Plot random numbers

```python
x=np.arange(50)# x-axis is numbers from 0 to 49
y=np.random.random(50) # y-axis are 50 random numbers
plt.plot(x,y, linestyle="--") # We the specify the style of plotting
```

### 4) Plot multiple lines in same graph 

```python
# First plot
x=np.arange(50)# x-axis is numbers from 0 to 49
y=np.random.random(50) # y-axis are 50 random numbers
plt.plot(x,y, linestyle="--")

# Second plot
x1=np.arange(50)
y1=np.random.random(50)
plt.plot(x1,y1)
```

### 5) Having 2 plots separately (one below another)

```python
# First plot
plt.figure()
x=np.arange(50)# x-axis is numbers from 0 to 49
y=np.random.random(50) # y-axis are 50 random numbers
plt.plot(x,y, linestyle="--")

# Second plot
plt.figure()
x1=np.arange(50)
y1=np.random.random(50)
plt.plot(x1,y1)
```

### 6) Scatter plot
A graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
x=np.arange(50)
y=np.random.random(50)
plt.scatter(x,y)
```

### 7) Combine plot and scatter

```python
x=np.arange(50)
y=np.random.random(50)
plt.plot(x,y)
plt.scatter(x,y,marker="*") # We can even change the marker
```

## 8) Histogram 

A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range. A histogram displays the shape and spread of continuous sample data.

```python
values=np.random.randint(0,50,20) # Displays 20 integer random numbers between 0 to 50
plt.hist(values)
```

### 9) Histogram with bins

```python
# Generate 20 random integer numbers between 0 and 11
values=np.random.randint(0,11,20)
print(values)


bins=np.linspace(0,10,6) # Generates 6 numbers between 0 and 10 which are equally spaced. 
print(bins) # bins are used to divide a histogram into equal parts. In our case, each bin is of size 2.


plt.hist(values,bins,width=1.9) # # change width of bars abit, so we can see space in between
```

# Object-oriented approach (preferred approach)

The object-oriented approach is more powerful and should be used when you want more control of your visualization. Instead of letting matplotlib handle what figure and axes that is the current one, we assign the figure and axes objects to variables and apply functions via methods on them.

### 1) Figures and axes objects

```python
f, ax = plt.subplots() # f is the canvas object, can contain several plots
print(f)
print(ax)

ax.plot([11,22,33],[1,10,5])
ax.hist(np.random.randint(10,33,15))
```

### 2) Several subplots on the same Figure

```python
f, ax = plt.subplots(nrows=2,ncols=2) # Craete 4 subplots
```

```python
f, ax = plt.subplots(nrows=2,ncols=2) # Craete 4 subplots

x=np.arange(100)
y=np.random.random(100)

print(x)
print('\n')
print(y)
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].hist(values)
ax[1,1].barh(x,np.abs(y))
```

### 3) Plot styling

```python
fig,ax=plt.subplots()

# First plot
x=np.arange(25)# x-axis is numbers from 0 to 49
y=np.sin(x)
ax.plot(x,y,color="green" ,linestyle="--",label="sin wave")

# Second plot
x1=np.arange(25)
y1=np.cos(x1)
ax.plot(x1,y1,color="red",linestyle=":",label="cos wave")

ax.legend(loc=4) 
# legend is a list of the names for each of the series on the chart.
# location of legend is an integer.
# for text to be correct in the legend the plots need a label


ax.set_xlabel("X axis")  # label of x-axis
ax.set_ylabel("y axis") # label of y-axis

ax.set_title("Sin and cos wave") # The heading of a graph
```

### 4) Set limits and annotate values in the plot

```python
x=np.arange(0,10,2) # array([0, 2, 4, 6, 8])
y=np.array([2,1,4,3,5])

fig,ax=plt.subplots()

ax.plot(x,y)

for i in zip(x,y):
    ax.annotate(s=str(i)  ,xy=i  )
```

### 5) plot sharing x-axis 

```python
# We will create two plots which share same x-axis.
x=np.arange(0,10,2)
y1=np.sin(x)
y2=np.cos(x)

f,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,sharex=True) # We are sharing x axis so, sharex=True

ax1.plot(x,y1)
ax2.plot(x,y2)
```

### 6) Saving a figure

```python
f.savefig('fig.png')
```
